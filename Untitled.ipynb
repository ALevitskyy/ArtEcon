```python
import pandas as pd
```

```python
date="2016-03-25 10:50:00"
currency="1"
number=10000
skipping=2
```

```python
index=pd.read_pickle("index.pkl")
index=index.index.get_loc(pd.to_datetime(date),
                                                method='nearest')
index=int(currency)*260885+index
index+=skipping
file_number=int(index/2500)*2500
leftover=index%2500
pandas_params=dict(index_col=0,
                           names=["coin","high","low","open","close"],
                           skiprows=0,
                           header=0,
                           nrows=leftover)
if leftover<number:
    data=pd.read_csv("cur_data/curdata_"+str(file_number)+
                                  ".csv",**pandas_params)
else:
    pandas_params["nrows"]=number
    pandas_params["skiprows"]=leftover-number
    data=pd.read_csv("cur_data/curdata_"+str(file_number)+
                             ".csv",**pandas_params)
candles_left=number-leftover
while candles_left>2500:
    pandas_params["nrows"]=2500
    file_number-=2500
    data2=pd.read_csv("cur_data/curdata_"+str(file_number)+
                            ".csv",**pandas_params)
            
    data=pd.concat([data2,data])
    candles_left-=2500
if candles_left>0:
    pandas_params["skiprows"]=2500-candles_left
    pandas_params["nrows"]=candles_left
    file_number-=2500
    data2=pd.read_csv("cur_data/curdata_"+str(file_number)+
                             ".csv",**pandas_params)
    data=pd.concat([data2,data])
data["date"]=data.index
```

```python
data=data.resample('5T').agg({'open': "first",
                                 'high': "max",
                                'low': "min",
                                'close': "last",
                                'date': "first"})
data=data.dropna(axis=0, how='any')
```
